```python
import joblib
import os
import pathlib
```

```python
PHY_DOMAIN = "WILDE_13"
CWD = os.getcwd()
DSD = pathlib.Path(os.path.join(CWD, "EXPERIMENT", PHY_DOMAIN))

for file in [ pathlib.Path(os.path.join(DSD, x)) for x in os.listdir(DSD) if "_data" in str(x)]:
    pac = joblib.load(file)
    print(len(pac))
    for i, p in enumerate(pac):
        print(f"[ FILE ]: {str(file)}")
        print(f"[ PAC ][ {str(i)} ]")
        for key in p.DATA.keys():
            print(f"{key}: {str(p.DATA[key])[:100]}")
```

```python
PHY_DOMAIN = "WILDE_13"
CWD = os.getcwd()
DSD = pathlib.Path(os.path.join(CWD, "EXPERIMENT", PHY_DOMAIN))

for file in [ pathlib.Path(os.path.join(DSD, x)) for x in os.listdir(DSD) if "_meta" in str(x)]:
    meta = joblib.load(file)
    print(f"[ FILE ]: {str(file)}")
    for key in meta.keys():
        print(f"{key}: {str(meta[key])[:100]}")
```
