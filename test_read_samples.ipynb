```python
import joblib
import os
import pathlib

from wifisidechannels.units.txbf import TxBf

N_SAMPLES       = 32
RANDOM_SPACING  = False
RANDOM_S        = False

SPACING = "random_spacing" if RANDOM_SPACING else "even_spacing"
SELECT  = "random_select" if RANDOM_S else "mean"

PHY_DOMAIN = ""
CWD = os.getcwd()
DSD = pathlib.Path(os.path.join(CWD, "DATASETS", SPACING, SELECT, str(N_SAMPLES)), PHY_DOMAIN)
UNIT = TxBf()

for i, file in enumerate(sorted([ pathlib.Path(os.path.join(folder, x)) for folder,_,y in os.walk(DSD) for x in y if ".dump" in x and "meta" not in str(x).lower()])):
    pac = joblib.load(file)
    print(i, len(pac), file)
    UNIT.plot_feedback_hist(
        packets=pac,
        plot=True,
        save_file=pathlib.Path(os.path.join(file.parents[0], file.stem + "_spectogram.png"))
        )
```
